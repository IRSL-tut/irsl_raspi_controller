## Raspbery piへの接続

````python
# 接続インスタンス作成
rpc = RPIController('AssembleRobot', 'ros_settings.yaml')
````

````python
# 設定ファイル送信
rpc.send_settings('robot_sensor.yaml', 'dynamixel_config.yaml', 'controller_config.yaml')
rpc.start_robot()
````

## RobotInterface立ち上げ

````python
ri = RobotInterface('robot_interface.yaml')
````

## RIの活用例

````python
ri.sendAngleVector([-1, 1], 4)
````

````python
tof_data = ri.waitNextData('TOFsensor0', timeout=1.0, clear=True)
print(tof_data)
````
